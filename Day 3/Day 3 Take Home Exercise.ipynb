```python
from tkinter import *
import tkinter.font as tkFont
```

```python
statelist = ['Andaman and Nicobar',
 'Andhra Pradesh',
 'Arunachal Pradesh',
 'Assam',
 'Bihar',
 'Chandigarh',
 'Chhattisgarh',
 'Dadra and Nagar Haveli',
 'Delhi',
 'Goa',
 'Gujarat',
 'Haryana',
 'Himachal Pradesh',
 'Jammu and Kashmir',
 'Jharkhand',
 'Karnataka',
 'Kerala',
 'Ladakh',
 'Lakshadweep',
 'Madhya Pradesh',
 'Maharashtra',
 'Manipur',
 'Meghalaya',
 'Mizoram',
 'Nagaland',
 'Odisha',
 'Puducherry',
 'Punjab',
 'Rajasthan',
 'Sikkim',
 'Tamil Nadu',
 'Telengana',
 'Tripura',
 'Uttar Pradesh',
 'Uttarakhand',
 'West Bengal'] 

#IMR function to calculate IMR value
def IMR():
    valinp.delete(first = 0, last = 100)
    inf = infinp.get()
    inf = int(inf)
    decr = decinp.get()
    decr = int(decr)
    state1 = variable.get()
    res = ((decr / inf) * 1000)
    mystr = "IMR value of " + state1 + " is " + str(res)
    valinp.insert(0, mystr) 
    infinp.delete(first = 1, last = 100)
    decinp.delete(first = 1, last = 100)

#CMR function to calculate CMR value
def CMR():
    valinp.delete(first = 1, last = 100)
    reck = recinp.get()
    reck = int(reck)
    decr = decinp.get()
    decr = int(decr)
    state1 = variable.get()
    res = ((decr / (reck + decr)) * 100)
    mystr = "CMR value of " + state1 + " is " + str(res)
    valinp.insert(0, mystr)
    recinp.delete(first = 1, last = 100)
    decinp.delete(first = 1, last = 100)

#Tkinter window is initialized    
window = Tk()

#Title for window
window.title("Pandemic Analytics")

#Size of window
window.geometry('600x400')
window.configure(bg = "lightgray")

#Font declaration
fontStyle1 = tkFont.Font(family = "Times", size = 40)
fontStyle2 = tkFont.Font(family = "Times", size = 20)
fontStyle3 = tkFont.Font(family = "Times", size = 20)

#Corona image 
corphoto = PhotoImage(file = "COVID-19.png")
corphotoin = corphoto.subsample(20, 20)
icon = Button(window, image = corphotoin)
icon.place(x = 50, y = 35, anchor = "center")

#Title of analyzer
titlepan = Label(window, text = "Pandemic Analyzer", font = fontStyle1, fg = "blue4")
titlepan.place(x = 350, y = 35, anchor = "center")

#State list
state = Label(window, text = "Select State:", font = fontStyle2)
state.place(x = 150, y = 100, anchor = "center")
variable = StringVar(window)
variable.set(statelist[0])
states = OptionMenu(window, variable, *statelist)
states.place(x = 400, y = 100, anchor = "center")

#Infected number label & Entry
infected = Label(window, text = "Infected:", font = fontStyle2, fg = "yellow")
infected.place(x = 150, y = 150, anchor = "center")
infinp = Entry(window, bd = 1, bg = "yellow")
infinp.place(x = 400, y = 150, anchor = "center")

#Recovered number label & Entry
recovered = Label(window, text = "Recovered:", font = fontStyle2, fg = "green")
recovered.place(x = 150, y = 200, anchor = "center")
recinp = Entry(window, bd = 1, bg = "green")
recinp.place(x = 400, y = 200, anchor = "center")

#Deceased number label & Entry
deceased = Label(window, text = "Deceased:", font = fontStyle2, fg = "red")
deceased.place(x = 150, y = 250, anchor = "center")
decinp = Entry(window, bd = 1, bg = "red")
decinp.place(x = 400, y = 250, anchor = "center")

#IMR & CMR Buttons
imrbtn = Button(window, text = "IMR", width = 5, height = 1, command = IMR)
imrbtn.place(x = 350, y = 300, anchor = "center")
cmrbtn = Button(window, text = "CMR", width = 5, height = 1, command = CMR)
cmrbtn.place(x = 450, y = 300, anchor = "center")

#Show result
value = Label(window, text = "Value:", font = fontStyle2, fg = "blue")
value.place(x = 150, y = 350, anchor = "center")
valinp = Entry(window, bd = 1, width = 45)
valinp.place(x = 400, y = 350, anchor = "center")
window.mainloop()
```
